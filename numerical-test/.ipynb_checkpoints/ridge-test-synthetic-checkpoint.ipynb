```python
# --- import packages ---
import sys 
sys.path.append("..") 
from source import utils,env,main,test
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
# import random
# import math
# import time
import gc
```

### Set test settings

```python
# x-axis

# ICML submission
# gamma_set = [0.8,0.74,0.7,0.64,0.56,0.5,0.44,0.4,0.36]

# Increase gamma range
# gamma_set = [0.8,0.7,0.56,0.4]
# gamma_set = [2,1.8,1.6,1.4,1.2,1]

# retest
gamma_set = [2.2,2,1.8,1.6,1.4]

# Fix the general setting
test_settings = {
    "time_lim":15 * 60,"ite_lim":None,"gap_lim":0.01,
    "num":50,"dim":300,"rho":0.5,"sparsity":10,"SNR":3.5,
    "inc_max_num":10,"inc_max_len":2,"exc_max_num":290,"exc_max_len":3, "tail_len":1,"gap_len":1,
    "method": 'conic'}

total_instance = 10
```

```python
# start testing

# store bad instance
bad_ins = []

# record attributes
all_att = []

# record sol time 
all_data_AA,all_data_Gen = [],[]

# record finding cuts time
all_find_cuts_AA,all_find_cuts_Gen = [],[]

# record cuts
all_support_AA,all_zero_AA = [],[]
all_inc_Gen, all_exc_Gen = [],[]

for gamma_ in gamma_set:
    print('=====')

    # record attributes for each gamma
    record_att = []

    # record sol time for each gamma
    record_AA,record_Gen = [],[]

    # record finding cuts time for each gamma
    record_find_cuts_AA,record_find_cuts_Gen = [],[]

    # record cuts for each gamma
    record_support_AA, record_zero_AA = [],[]
    record_inc_Gen, record_exc_Gen = [],[]
    for _ in range(total_instance):
        time_log, cut_log, att_log, test_X, test_Y = test.single_ins_test(progress_log=True,test_alg=['Gen'], \
                                                            gamma_ = gamma_,**test_settings)

        # record bad instance
        if att_log['Gen status'] != 2:
            bad_ins.append((test_X,test_Y,gamma_,test_settings['sparsity']))
        else:
            pass
        del test_X
        del test_Y
        gc.collect
        
        record_AA.append(time_log['AA'])
        record_Gen.append(time_log['Gen'])

        record_att.append(att_log)
        # record_att_Gen.append(obj_log['Gen'])

        record_find_cuts_AA.append(time_log['AA find cuts'])
        record_find_cuts_Gen.append(time_log['Gen find cuts'])

        record_support_AA.append(cut_log['AA_support'])
        record_zero_AA.append(cut_log['AA_zero'])

        record_inc_Gen.append(cut_log['Gen_inclusive'])
        record_exc_Gen.append(cut_log['Gen_exclusive'])

        # delete the variable
        del time_log
        del cut_log
        del att_log
        gc.collect()
        

    
    all_att.append(record_att)
    # all_obj_Gen.append(np.array(record_obj_Gen))
    
    all_data_AA.append(np.array(record_AA))
    all_data_Gen.append(np.array(record_Gen))

    all_find_cuts_AA.append(np.array(record_find_cuts_AA))
    all_find_cuts_Gen.append(np.array(record_find_cuts_Gen))

    all_support_AA.append(record_support_AA)
    all_zero_AA.append(record_zero_AA)

    all_inc_Gen.append(record_inc_Gen)
    all_exc_Gen.append(record_exc_Gen)
    print('One gamma done',gamma_)
    print('=====')
```

```python
# check the bad instance
check = True
save = True

if check == True:
    print('bad ins num:', len(bad_ins))
    print(bad_ins)
    

if save == True:
    save_dic = {}
    save_dic['X'] = []
    save_dic['Y'] = []
    save_dic['gamma'] = []
    save_dic['sparsity'] = []

    if bad_ins !=[]:
        for i in range(len(bad_ins)):
            ins_tuple = bad_ins[i]
            X,Y,gma,spar = ins_tuple[0],ins_tuple[1],ins_tuple[2],ins_tuple[3]
            save_dic['X'].append(X)
            save_dic['Y'].append(Y)
            save_dic['gamma'].append(gma)
            save_dic['sparsity'].append(spar)
    
        # save bad instance
        file_name = 'k_' + str(test_settings['sparsity']) + '_SNR_'+ str(test_settings['SNR']) + '_bad' + '.npy'
        np.save(file_name,save_dic)
```

### Save data

```python
# save data
save_dict = dict()

# save global settings
save_dict['test_settings'] = test_settings
save_dict['gamma_set'] = gamma_set
save_dict['total_instance'] = total_instance

save_dict['attributes'] = all_att
save_dict['AA_time'] = all_data_AA
save_dict['Gen_time'] = all_data_Gen

save_dict['AA find cuts'] = all_find_cuts_AA
save_dict['Gen find cuts'] = all_find_cuts_Gen

save_dict['AA_support'],save_dict['AA_zero'] = all_support_AA,all_zero_AA
save_dict['Gen_inc'],save_dict['Gen_exc'] = all_inc_Gen,all_exc_Gen

# save
file_name = 'k_' + str(test_settings['sparsity']) + '_SNR_'+ str(test_settings['SNR']) \
            + '_total_instance_' + str(total_instance) + '.npy'
np.save(file_name,save_dict)
```

### Objective ratio

```python
# Retrieve the objective
AA_obj_list = []
Gen_obj_list = []

for i in range(len(gamma_set)):
    AA_obj_gamma = []
    Gen_obj_gamma = []
    for j in range(total_instance):
        AA_obj_gamma.append(all_att[i][j]['AA obj'])
        Gen_obj_gamma.append(all_att[i][j]['Gen obj'])

    AA_obj_list.append(AA_obj_gamma)
    Gen_obj_list.append(Gen_obj_gamma)

# get the objective ratio
ratio_list = []
for i in range(len(gamma_set)):
    ratio_gamma = []
    for j in range(total_instance):
        ratio_gamma.append( Gen_obj_list[i][j] / AA_obj_list[i][j])

    ratio_list.append(ratio_gamma)

# Plot settings
sns.set_style('whitegrid')
font1 = {'weight' : 'bold', 
         'size'   : 25}
fig = plt.figure(figsize=(15,10))

# Draw Average 
line_avg_gap, = plt.plot(gamma_set,np.mean(ratio_list,axis = 1),linestyle="-",color="red",marker="o",linewidth=3.5,markersize = 13)

# Fill with standard deviation
avg = np.mean(ratio_list,axis = 1)
std=np.std(ratio_list,axis=1)
r1 = list(map(lambda x: x[0]-x[1], zip(avg, std)))
r2 = list(map(lambda x: x[0]+x[1], zip(avg, std)))
plt.fill_between(gamma_set,r1,r2,color = 'orange',alpha = 0.2)

ax = plt.gca() 
ax.spines['top'].set_color('black')  
ax.spines['right'].set_color('black')
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_color('black')
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)

plt.xticks(fontsize=30,fontweight = 'bold')
plt.yticks(fontsize=30,fontweight = 'bold')
plt.xlabel('gamma',fontsize=35,fontweight = 'bold')
plt.ylabel('Objective ratio',fontsize=35,labelpad = 20,fontweight = 'bold')
plt.title('sparsity = '+str(test_settings['sparsity']) + \
          '; rho = ' + str(test_settings['rho']) + \
          '; SNR = ' + str(test_settings['SNR']) \
          ,fontsize=35,pad = 20,fontweight = 'bold')
# plt.legend([line_avg_gap],['time'],loc = 'upper left',prop = font1)

# file_name = './dim_100_50_fig/'+ \
#         'time_gap_'+ \
#         'k_' + str(test_settings['sparsity'])+ \
#         '_SNR_' + str(test_settings['SNR'])+ \
#         '_total_instance_' + \
#         str(total_instance)+'.pdf'
# plt.savefig(file_name,bbox_inches = 'tight')
```

### MIPGap Ratio

```python
# Retrieve the MIPGap
AA_gap_list = []
Gen_gap_list = []

for i in range(len(gamma_set)):
    AA_gap_gamma = []
    Gen_gap_gamma = []
    for j in range(total_instance):
        AA_gap_gamma.append(all_att[i][j]['AA MIPGap'])
        Gen_gap_gamma.append(all_att[i][j]['Gen MIPGap'])

    AA_gap_list.append(AA_gap_gamma)
    Gen_gap_list.append(Gen_gap_gamma)

# get the MIPGap ratio
ratio_list = []
for i in range(len(gamma_set)):
    ratio_gamma = []
    for j in range(total_instance):
        
        # AA solved to optimal
        if AA_gap_list[i][j] <= 0.01:
            
            # Gen solved to optimal
            if Gen_gap_list[i][j] <= 0.01:
                ratio_gamma.append(1)
            # Gen not solved to optimal
            else:
                ratio_gamma.append(999)
        else:
            ratio_gamma.append( Gen_gap_list[i][j] / AA_gap_list[i][j])

    ratio_list.append(ratio_gamma)

# Plot settings
sns.set_style('whitegrid')
font1 = {'weight' : 'bold', 
         'size'   : 25}
fig = plt.figure(figsize=(15,10))

# Draw Average 
line_avg_gap, = plt.plot(gamma_set,np.mean(ratio_list,axis = 1),linestyle="-",color="red",marker="o",linewidth=3.5,markersize = 13)

# Fill with standard deviation
avg = np.mean(ratio_list,axis = 1)
std=np.std(ratio_list,axis=1)
r1 = list(map(lambda x: x[0]-x[1], zip(avg, std)))
r2 = list(map(lambda x: x[0]+x[1], zip(avg, std)))
plt.fill_between(gamma_set,r1,r2,color = 'orange',alpha = 0.2)

ax = plt.gca() 
ax.spines['top'].set_color('black')  
ax.spines['right'].set_color('black')
ax.spines['left'].set_color('black')
ax.spines['bottom'].set_color('black')
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)

plt.xticks(fontsize=30,fontweight = 'bold')
plt.yticks(fontsize=30,fontweight = 'bold')
plt.xlabel('gamma',fontsize=35,fontweight = 'bold')
plt.ylabel('MIPGap ratio',fontsize=35,labelpad = 20,fontweight = 'bold')
plt.title('sparsity = '+str(test_settings['sparsity']) + \
          '; rho = ' + str(test_settings['rho']) + \
          '; SNR = ' + str(test_settings['SNR']) \
          ,fontsize=35,pad = 20,fontweight = 'bold')
# plt.legend([line_avg_gap],['time'],loc = 'upper left',prop = font1)

# file_name = './dim_100_50_fig/'+ \
#         'time_gap_'+ \
#         'k_' + str(test_settings['sparsity'])+ \
#         '_SNR_' + str(test_settings['SNR'])+ \
#         '_total_instance_' + \
#         str(total_instance)+'.pdf'
# plt.savefig(file_name,bbox_inches = 'tight')
```

### Cut characteristic table

```python
# Draw the comparison table
df = pd.DataFrame(columns = [str(gamma) for gamma in gamma_set
                            ])

good_minus_num_list = []
for i in range(len(gamma_set)):
    time_result = relative_time_gap[i]
    good_num = 0
    for result in time_result:
        if result > 0:
            good_num += 1

    minus_num = total_instance - good_num
    good_minus_num_list.append((good_num,minus_num))

df.loc[len(df)] = good_minus_num_list
df.loc[len(df)] = np.mean(relative_time_gap,axis = 1)

df.index = ['(Gen,AA)','relative time gap']
df
```

```python
# Draw the cut characteristic table 
df = pd.DataFrame(columns = [str(gamma) for gamma in gamma_set
                            ])

avg_support_AA, avg_zero_AA = [],[]
num_no_support_AA, num_no_zero_AA = [],[]

for i in range(len(gamma_set)):
    avg_support_AA.append(np.mean([len(j) for j in all_support_AA[i]]))
    avg_zero_AA.append(np.mean([len(j) for j in all_zero_AA[i]]))
    num_no_support_AA.append(len([j for j in all_support_AA[i] if j == []]))
    num_no_zero_AA.append(len([j for j in all_zero_AA[i] if j == []]))

avg_num_inc,avg_num_exc = [],[]
avg_len_inc,avg_len_exc = [],[]
num_no_inc,num_no_exc = [],[]

for i in range(len(gamma_set)):
    avg_num_inc.append(np.mean(
                            [len(j) for j in all_inc_Gen[i] if j!=[[]] ] + \
                                [0 for j in all_inc_Gen[i] if j==[[]] ])
                                  )
    avg_num_exc.append(np.mean(
                            [len(j) for j in all_exc_Gen[i] if j!=[[]] ] + \
                                [0 for j in all_exc_Gen[i] if j==[[]] ])
                                  )
    fix_gamma_no_inc, fix_gamma_no_exc = 0,0
    fix_gamma_inc_len, fix_gamma_exc_len = [],[]
    for j in range(total_instance):
        inc_set , exc_set = all_inc_Gen[i][j] , all_exc_Gen[i][j]
        if inc_set == [[]]:
            fix_gamma_no_inc += 1
        else:
            fix_gamma_inc_len.append(np.mean([len(cut) for cut in inc_set]))
        if exc_set == [[]]:
            fix_gamma_no_exc += 1
        else:
            fix_gamma_exc_len.append(np.mean([len(cut) for cut in exc_set]))
            
    num_no_inc.append(fix_gamma_no_inc)
    num_no_exc.append(fix_gamma_no_exc)

    avg_len_inc.append(np.mean(fix_gamma_inc_len))
    avg_len_exc.append(np.mean(fix_gamma_exc_len))


# first add AA
df.loc[len(df)] = avg_support_AA
df.loc[len(df)] = avg_zero_AA
df.loc[len(df)] = num_no_support_AA
df.loc[len(df)] = num_no_zero_AA

# next add Gen
df.loc[len(df)] = avg_num_inc
df.loc[len(df)] = avg_num_exc
df.loc[len(df)] = avg_len_inc
df.loc[len(df)] = avg_len_exc
df.loc[len(df)] = num_no_inc
df.loc[len(df)] = num_no_exc

    

df.index = ['average support that can be fixed (AA)','average zero that can be fixed (AA)',
            'num of no support fixed instances (AA)','num of no zero fixed instances (AA)',
            'average num of inc cuts (Gen)','average num of exc cuts (Gen)',
            'average len of inc cut (Gen)','average len of exc cut (Gen)',
            'num of no inc cut instances (Gen)','num of no exc cut instances (Gen)']

df
```
